```python
import os
from dotenv import load_dotenv
from polygon import RESTClient
import json
import pandas as pd

load_dotenv() 


API_KEY = os.getenv("API_KEY")
client = RESTClient(api_key=API_KEY)
```

```python
ticker = "AMZN"

# List Aggregates (Bars)
aggs = []
for a in client.list_aggs(ticker=ticker, multiplier=1, timespan="hour", from_="2023-01-01", to="2024-01-01", limit=50000):
    aggs.append(a)

df = pd.DataFrame(aggs)
```

```python
df['Date'] = df['timestamp'].apply(lambda x: pd.to_datetime(x*1000000))
# Assuming 'df' is your DataFrame and 'column_to_drop' is the name of the column you want to drop
df = df.drop(columns=['otc'])
df = df.drop(columns=['timestamp'])
df
```
